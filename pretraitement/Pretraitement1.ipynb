#### Le but de ce carnet est de nettoyer le dataset.

### 1- Nous allons commencer par supprimer les doublons: les fichiers se terminant par vsx ou x est un entier. Par la suite, nous allons afficher le nombre de fichiers supprimé.

```python
import os
import re
```

```python
root_dir = r"C:\Users\Christian\Desktop\YembaTones\YembaTones An Annotated Dataset for Tonal and Syllabic Analysis of the Yemba Language\Yemba_Dataset\audios"

# Detection des expressions régulières pour détecter les fichiers "vsX"
pattern = re.compile(r"vs\d+\.(wav|TextGrid)$", re.IGNORECASE)

# Compteurs
deleted_files = []

# Parcours récursif
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if pattern.search(filename):
            file_path = os.path.join(dirpath, filename)
            try:
                os.remove(file_path)
                deleted_files.append(file_path)
                print(f"Supprimé : {file_path}")
            except Exception as e:
                print(f"Erreur lors de la suppression de {file_path} : {e}")

print(f"Nombre total de fichiers supprimés : {len(deleted_files)}")
```

#### Suppression manuelle du dossiers speaker_9_group_28vsprobleme

```python
import os
import pandas as pd
import re

# 1. Charger le dictionnaire
dict_df = pd.read_excel(r"C:\Users\Christian\Desktop\YembaTones\YembaTones An Annotated Dataset for Tonal and Syllabic Analysis of the Yemba Language\Yemba_Dataset\metadata\isolated_words_dictionary.xlsx")

# 2. Initialiser la liste pour les lignes du CSV
rows = []

# 3. Racine des fichiers audio
root_audio_dir = r"C:\Users\Christian\Desktop\YembaTones\YembaTones An Annotated Dataset for Tonal and Syllabic Analysis of the Yemba Language\Yemba_Dataset\audios"

# 4. Parcours récursif des fichiers audio
for dirpath, dirnames, filenames in os.walk(root_audio_dir):
    for file in filenames:
        if file.endswith(".wav") and "vs" not in file:  # on ignore les variantes vs
            match = re.match(r"spkr_(\d+)_group_(\d+)_statement_(\d+).wav", file)
            if match:
                speaker_id = int(match.group(1))
                group_id = int(match.group(2))
                statement_id = int(match.group(3))

                # Cherche la ligne correspondante dans le dictionnaire (WordID == statement_id)
                dict_row = dict_df[dict_df["WordId"] == statement_id]
                if not dict_row.empty:
                    dict_row = dict_row.iloc[0]
                    row = {
                        "WordID": statement_id,
                        "MotYemba": dict_row["Yemba"],
                        "Syllabe 1": dict_row["Syllabe 1"],
                        "Ton 1": dict_row["Tone 1"],
                        "Syllabe 2": dict_row["Syllabe 2"],
                        "Ton 2": dict_row["Tone 2"],
                        "Syllabe 3": dict_row["Syllabe 3"],
                        "Ton 3": dict_row["Tone 3"],
                        "Speaker": speaker_id,
                        "Group": group_id
                    }
                    rows.append(row)

# 5. Convertir en DataFrame
final_df = pd.DataFrame(rows)

# 6. Enregistrer en CSV
output_path = "yemba_dataset_structured.csv"
final_df.to_csv(output_path, index=False)

print(f"Fichier CSV généré : {output_path}")
```
